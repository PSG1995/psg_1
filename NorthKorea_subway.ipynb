```python
import folium
import pandas as pd
```

```python
dat = pd.read_csv('북한지도_철도_지하철역_EPSG4326_EUC_KR.csv',encoding='EUCKR')
dat
```

```python
dat.columns
```

```python
dat_tmp = dat.loc[0:, ['지명','Y좌표','X좌표']]
dat_tmp.columns = ['name', '위도', '경도']
dat_tmp
```

```python
## 북한 선택
NorthK = folium.Map(location=[40.081519, 124.499307], zoom_start=10)
NorthK
```

```python
dat_tmp.shape
```

```python
des = dat_tmp['name']
dat_p1 = list(dat_tmp['위도'])
dat_p2 = list(dat_tmp['경도'])
```

```python
dat_xy = list(zip(dat_p2,dat_p1))
dat_xy[0:3]
```

```python
NorthK = folium.Map(location=[40.081519, 124.499307], zoom_start=10)
for name, loc in zip(des, dat_xy):
  folium.Marker(location=loc,
                popup=name,
                tooltip=name,
                icon=folium.Icon(color='blue',icon='info-sign')).add_to(NorthK)
NorthK.save('northkorea_sobway.html')
NorthK
```
